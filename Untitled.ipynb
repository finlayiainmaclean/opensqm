```python
import openmm as mm
from openmm import app, unit
import numpy as np
import random

class FocusedSimulationManager:
    def __init__(self, pdb_path, ff_path, ligand_indices, radius=0.8, k=10000.0, temp=300.0, pressure=1.0):
        self.ligand_indices = ligand_indices
        self.radius_nm = radius
        self.k_val = k 
        self.temp = temp * unit.kelvin
        self.pressure = pressure * unit.atmosphere
        
        pdb = app.PDBFile(pdb_path)
        ff = app.ForceField(ff_path)
        
        # 1. Setup GLOBAL Context (NPT Ensemble)
        self.system_global = ff.createSystem(pdb.topology, nonbondedMethod=app.PME, 
                                            nonbondedCutoff=1.0*unit.nanometer, 
                                            constraints=app.HBonds)
        
        # ADD BAROSTAT TO GLOBAL ONLY
        self.barostat = mm.MonteCarloBarostat(self.pressure, self.temp, 25) # Update every 25 steps
        self.system_global.addForce(self.barostat)
        
        int_global = mm.LangevinMiddleIntegrator(self.temp, 1/unit.picosecond, 0.002*unit.picoseconds)
        int_global.setIntegrationForceGroups({0})

        
        self.sim_global = app.Simulation(pdb.topology, self.system_global, int_global)
        self.sim_global.context.setPositions(pdb.positions)
        self.sim_global.context.setVelocitiesToTemperature(self.temp)

        # 2. Setup LOCAL Context (NVT Ensemble - Optimized)
        # Local MD does NOT get a barostat.
        self.system_local = ff.createSystem(pdb.topology, nonbondedMethod=app.PME, 
                                           nonbondedCutoff=1.0*unit.nanometer, 
                                           constraints=app.HBonds)
        
        initial_center = random.choice(self.ligand_indices)
        self._setup_local_constraints_and_forces(initial_center, pdb.positions)
        
        int_local = mm.LangevinMiddleIntegrator(self.temp, 1/unit.picosecond, 0.002*unit.picoseconds)
        int_local.setIntegrationForceGroups({0})
        self.sim_local = app.Simulation(pdb.topology, self.system_local, int_local)

    def update_local_parameters(self):
        """
        The 'Resampling' part: 
        1. Pick a random ligand atom to be the new center 'c'.
        2. Identify atoms within the radius of this new 'c'.
        3. Update the local context parameters.
        """
        # 1. Randomly choose the center index for this block
        self.current_center_idx = random.choice(self.ligand_indices)
        
        # 2. Get current positions from the Global Simulation
        state = self.sim_global.context.getState(getPositions=True)
        pos = state.getPositions(asNumpy=True).value_in_unit(unit.nanometer)
        center_pos = pos[self.current_center_idx]
        
        # 3. Update the 'auxiliary variables' (restraint origin) in the Local Context
        self.sim_local.context.setParameter("x0", center_pos[0])
        self.sim_local.context.setParameter("y0", center_pos[1])
        self.sim_local.context.setParameter("z0", center_pos[2])
        
        # Note: In a fully dynamic implementation, you would also update the 
        # Interaction Groups here if the ligand moves significantly.
        # However, updating the restraint center (x0, y0, z0) is the 
        # mathematically required step to avoid bias.

    def _setup_local_constraints_and_forces(self, center_idx, positions):
        pos_array = positions.value_in_unit(unit.nanometer)
        center_pos = pos_array[center_idx]
        
        active_set = []
        for i in range(self.system_local.getNumParticles()):
            dist = np.linalg.norm(pos_array[i] - center_pos)
            if dist < self.radius_nm:
                active_set.append(i)
                m = self.system_global.getParticleMass(i)
                self.system_local.setParticleMass(i, m)
            else:
                self.system_local.setParticleMass(i, 0.0 * unit.dalton)
        
        for force in self.system_local.getForces():
            if isinstance(force, mm.NonbondedForce):
                force.addInteractionGroup(active_set, range(self.system_local.getNumParticles()))
        
        fb_expr = "0.25*k*step(r-r0)*(r-r0)^4; r=distance(x,y,z, x0,y0,z0)"
        restraint = mm.CustomExternalForce(fb_expr)
        restraint.addGlobalParameter("k", self.k_val * unit.kilojoule_per_mole/unit.nanometer**4)
        restraint.addGlobalParameter("r0", self.radius_nm * unit.nanometer)
        restraint.addGlobalParameter("x0", center_pos[0])
        restraint.addGlobalParameter("y0", center_pos[1])
        restraint.addGlobalParameter("z0", center_pos[2])
        
        for idx in active_set:
            if idx == center_idx:
                self.system_local.setParticleMass(idx, 0.0 * unit.dalton)
            else:
                restraint.addParticle(idx, [])
        self.system_local.addForce(restraint)

    def sync_state(self, source_sim, target_sim):
        """
        Copies positions, velocities, AND Periodic Box Vectors.
        Crucial because the Global Barostat changes the box size.
        """
        state = source_sim.context.getState(getPositions=True, getVelocities=True)
        target_sim.context.setPeriodicBoxVectors(*state.getPeriodicBoxVectors())
        target_sim.context.setPositions(state.getPositions())
        target_sim.context.setVelocities(state.getVelocities())

    def step_hybrid(self, cycles=10, global_steps=500, local_steps=1000):
        for cycle in range(cycles):
            # 1. GLOBAL STEP (NPT)
            print(f"Cycle {cycle}: Global MD (NPT)...")
            self.sim_global.step(global_steps)
            
            # 2. SYNC & RESAMPLE
            # We must pass the new box vectors from NPT global to NVT local
            self.sync_state(self.sim_global, self.sim_local)
            
            new_center_idx = random.choice(self.ligand_indices)
            curr_pos = self.sim_global.context.getState(getPositions=True).getPositions(asNumpy=True)
            c_pos = curr_pos[new_center_idx].value_in_unit(unit.nanometer)
            
            self.sim_local.context.setParameter("x0", c_pos[0])
            self.sim_local.context.setParameter("y0", c_pos[1])
            self.sim_local.context.setParameter("z0", c_pos[2])
            
            # 3. LOCAL STEP (NVT)
            print(f"Cycle {cycle}: Local Focused MD (NVT, Center: {new_center_idx})")
            self.sim_local.step(local_steps)
            
            # 4. SYNC BACK
            self.sync_state(self.sim_local, self.sim_global)
```

```python
def add_energy_probe(system, probe_index, group_id=31):
    """
    Adds a CustomNonbondedForce to the system that specifically calculates 
    the interaction energy of one atom with everything else.
    """
    # 1. Define the energy expression (LJ + Coulomb)
    # This mimics the standard NonbondedForce physics
    energy_expression = ("4*epsilon*((sigma/r)^12 - (sigma/r)^6) + (138.935456*q1*q2)/r;"
                         "sigma = 0.5*(sigma1+sigma2);"
                         "epsilon = sqrt(epsilon1*epsilon2)")
    
    probe_force = mm.CustomNonbondedForce(energy_expression)
    probe_force.addPerParticleParameter("q")
    probe_force.addPerParticleParameter("sigma")
    probe_force.addPerParticleParameter("epsilon")
    
    # 2. Extract parameters from the existing NonbondedForce
    nb_force = [f for f in system.getForces() if isinstance(f, mm.NonbondedForce)][0]
    for i in range(system.getNumParticles()):
        q, sigma, epsilon = nb_force.getParticleParameters(i)
        probe_force.addParticle([q, sigma, epsilon])
        
    # 3. Define the Interaction Group: (Probe Atom) vs (All Other Atoms)
    # This is the 'Force Filtering' logic from the C++ reference
    probe_force.addInteractionGroup([probe_index], range(system.getNumParticles()))
    
    # 4. Set the Force Group so we can query this energy specifically
    probe_force.setForceGroup(group_id)
    probe_force.setNonbondedMethod(mm.CustomNonbondedForce.CutoffPeriodic)
    probe_force.setCutoffDistance(nb_force.getCutoffDistance())
    
    system.addForce(probe_force)
    return group_id
```

```python
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt

def run_benchmark(manager, probe_idx, n_samples=2000):
    """
    manager: The FocusedSimulationManager we built previously
    probe_idx: The index of the atom to monitor
    """
    # Add energy probes to both systems
    add_energy_probe(manager.system_global, probe_idx, group_id=31)
    add_energy_probe(manager.system_local, probe_idx, group_id=31)

    manager.sim_global.getIntegrator().setIntegrationForceGroups({0})
    manager.sim_local.getIntegrator().setIntegrationForceGroups({0})
    
    # Re-initialize to register the new forces
    manager.sim_global.context.reinitialize(preserveState=True)
    manager.sim_local.context.reinitialize(preserveState=True)
    
    global_energies = []
    local_energies = []

    print("Sampling Global MD Ground Truth...")
    for _ in range(n_samples):
        manager.sim_global.step(250)
        state = manager.sim_global.context.getState(getEnergy=True, groups={31})
        energy = state.getPotentialEnergy().value_in_unit(unit.kilojoule_per_mole)
        global_energies.append(energy)

    print("Sampling Local Focused MD...")
    # Sync and choose a random center (Resampling Trick)
    manager.sync_state(manager.sim_global, manager.sim_local)
    
    for _ in range(n_samples):
        # Every few steps, run the Resampling Trick as per the paper
        if _ % 5 == 0:
            manager.update_local_parameters() 
            
        manager.sim_local.step(250)
        state = manager.sim_local.context.getState(getEnergy=True, groups={31})
        energy = state.getPotentialEnergy().value_in_unit(unit.kilojoule_per_mole)
        local_energies.append(energy)

    # Statistical Analysis
    d_stat, p_val = ks_2samp(global_energies, local_energies)
    
    print(f"\n--- BENCHMARK RESULTS ---")
    print(f"KS Statistic (D): {d_stat:.4f}")
    print(f"P-Value: {p_val:.4f}")
    
    # Visualization
    plt.hist(global_energies, bins=50, alpha=0.5, label='Global MD (Ground Truth)')
    plt.hist(local_energies, bins=50, alpha=0.5, label='Local Focused MD')
    plt.xlabel('Probe Interaction Energy (kJ/mol)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.title(f'Sampling Bias Validation (D={d_stat:.4f})')
    plt.show()

    return d_stat
```

```python
from concurrent.futures import ThreadPoolExecutor, TimeoutError
import time

def do_work():
    # Simulate a slow operation
    time.sleep(10)
    return "finished"

def run_with_timeout(fn, timeout):
    with ThreadPoolExecutor(max_workers=1) as executor:
        future = executor.submit(fn)
        try:
            return future.result(timeout=timeout)
        except TimeoutError:
            raise RuntimeError("Operation timed out")

# usage
result = run_with_timeout(do_work, 5.0)
print(result)
```
